```python
Combine = pd.DataFrame(columns = ['Apple'])
Combine['Apple']=Apple['Close']
Combine['Lv']=lv['Close']
Combine['UBonds']=bonds['Close']
Combine['GBonds']=bonds2['Close']
Combine['Gold']=golden['Close']
Combine['Silver']=silver['Close']
# Combine['Oil'] =oil['Close']
Combine['Dow']=DOW['Close']
Combine['SPX'] =SPX['Close']
Combine['AEX'] =AEX['Close']
# Combine['DAX'] =DAX['Close']
Combine['Bitcoin'] =bitcoin['Close']
Combine.dropna(inplace=True)
```

```python
pect_change = Combine.pct_change()
pect_change['UBonds']= Combine['UBonds'].diff()
pect_change['GBonds']= Combine['GBonds'].diff()
pect_change = pect_change.dropna()
pect_change = pect_change*-1
```

```python
pect_change.to_csv('Data1.csv')
```

```python
from sklearn.preprocessing import StandardScaler
after_normalize = StandardScaler().fit_transform(pect_change)
```

```python
normalised_data = pd.DataFrame(after_normalize,columns=pect_change.columns,index=pect_change.index)
```

```python
normalised_data.to_csv('Data.csv')
```

```python
from scipy.stats import t

def student_std_mean(df,v):
    port_mean = df.mean()
    student_vali = df.std() * np.sqrt((v-2)/v)
    return student_vali,port_mean
    
    

def Diff_studentt(df,v,invest=1):
    port_vali,port_mean=student_std_mean(df,v)
    dataframe=pd.DataFrame(columns = ['VaR_999','VaR_99','ES_999','ES_99'])
    dataframe['VaR_999']=np.arange(0,1,1)
    dataframe['VaR_999']=(port_mean +t.ppf(0.999,v)*port_vali)*invest
    dataframe['VaR_99'] = (port_mean +t.ppf(0.99, v)*port_vali)*invest
    dataframe['ES_999']= (port_mean + port_vali*(scipy.stats.t.pdf(scipy.stats.t.ppf(1-0.999, v),v)/(1-0.999))*((v+scipy.stats.t.ppf(1-0.999, v)**2)/(v-1)))*invest
    dataframe['ES_99'] = (port_mean + port_vali*(scipy.stats.t.pdf(scipy.stats.t.ppf(1-0.99,v),v)/(1-0.99))*((v+scipy.stats.t.ppf(1-0.99, v)**2)/(v-1)))*invest
    
    return dataframe
    
```
